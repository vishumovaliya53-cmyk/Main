```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')  # clean theme

print("Libraries loaded successfully!")
```

## 1. Load Dataset

```python
# Example dataset (Johns Hopkins or OWID format)
# Replace with your CSV file path
path = "sales_data_missing.csv"
df = pd.read_csv(path)
df.head()
```

## 2. Data Cleaning

```python
print("Missing Values:")
print(df.isnull().sum())

# Fill numeric missing values with 0
df = df.fillna(0)

print("Cleaning done!")
```

## 3. Summary Statistics

```python
df.describe()
```

```python
#For Null values
print(":-:-:-:-For The Null Values:-:-:-:-\n")
print(df.isnull().sum())
df.head()
```

```python
#fillna,isnull used to handles missing values
print(":-:-:-:-Handle the null values:-:-:-:-\n")
df.fillna(df.mean(numeric_only=True), inplace=True)
df.isnull().sum()
```

### Cleaned Data Saved

```python
#Here saved cleaned data
print(":-:-:-:-Saved:-:-:-:-\n")
df.to_csv("cleaned_student_performance.csv", index=False)
print(" Cleaned data saved successfully.")
```

## 4. Visualizations

### Bar chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sales_data_missing.csv")

df.columns = df.columns.astype(str).str.strip()

print(df.columns.tolist())

plt.figure(figsize=(8,5))

avg_profit = df.groupby('Region')['Profit'].mean()

sns.barplot(x=avg_profit.index, y=avg_profit.values, palette='coolwarm')
plt.title("Average Profit per Region")
plt.ylabel("Average Profit")
plt.show()
```

### Heat map

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("sales_data_missing.csv")

# Clean column names
df.columns = df.columns.astype(str).str.strip()

# Show columns
print(df.columns.tolist())

# Plot histogram
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Profit', kde=True)
plt.title("Distribution of Profit")
plt.xlabel("Profit")
plt.ylabel("Count")
plt.show()
```

### Countplot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("sales_data_missing.csv")

# Clean column names
df.columns = df.columns.astype(str).str.strip()

# Show columns
print(df.columns.tolist())

# Plot Countplot
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Region', palette='coolwarm')
plt.title("Count of Records per Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.show()
```

#### Line Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("sales_data_missing.csv")

# Clean column names
df.columns = df.columns.astype(str).str.strip()

# Line Chart
plt.figure(figsize=(8,5))
sns.lineplot(data=df, x='Year', y='Sales', marker='o')
plt.title("Sales Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

### Heatmap

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("sales_data_missing.csv")

# Clean column names
df.columns = df.columns.astype(str).str.strip()

# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

# Correlation matrix
corr = numeric_df.corr()

# Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(corr, annot=True, linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

### Scatter plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("sales_data_missing.csv")

# Clean column names
df.columns = df.columns.astype(str).str.strip()

# Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Sales', y='Profit')
plt.title("Sales vs Profit Scatter Plot")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```
