```python
import matplotlib.pyplot as plt
```

```python
x =["Science" , "Maths", "English"]
y = [90, 80, 70]

plt.bar(x, y)

plt.title("Subject wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show
```

```python
plt.bar(x, y, width=0.5)

plt.title("Subject wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")

plt.show()
```

```python
plt.bar(x, y, color = "lightblue", width=0.5)

plt.title("Subject wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")

plt.show() 
```
