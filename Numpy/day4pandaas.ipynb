### Seris

#### A One dimentional array like structure with indices as label

```python
s1 = pd.Series([10, 20, 30, 40])
print(s1)
```

```python
print(s1[0]) # Accessing first element
```

```python
print(s1[0:3]) # Slicing first three elements
```

```python
print(s1[0::2]) # Slicing with step size 2
```
