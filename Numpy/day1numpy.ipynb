# Indexing Of Array

```python
a2 = np.random.randint(3 , 10 ,(3,3)) # create a 3x3 array of random integers between 3 and 9
a2
```

```python
print(a2[0 , 1]) # access element at row 0, column 1
```

```python
print(a2[2 , 0]) # access element at row 2, column 0
```

```python
print(a2[2 , 1]) # access element at row 2, column 1
```

```python
print(a2[0 , 0]) # access element at row 0, column 0
```

# Slicing On Array

```python
a = np.arange(3 , 10) # create an array with values from 3 to 9
a
```

```python
a[0:3] # slice the array to get elements from index 0 to 2
```

```python
a[-1:-5:-1] # slice the array to get the last 4 elements in reverse order
```

```python
a[2:6]  # slice the array to get elements from index 2 to 5
```

```python
a[0:] # slice the array to get all elements from index 0 to the end
```

```python
a[0::2]  # slice the array to get every second element starting from index 0    
```

```python
a[::1] # slice the array to get all elements with a step of 1
```

```python
a2 = np.random.randint(1 , 10 , (3,3)) # create a 3x3 array of random integers between 1 and 9
a2
```

```python
a2 [0:2 , 2] # slice the array to get elements from rows 0 to 1 in column 2
```

```python
a2 [0: , 1] # slice the array to get all rows in column 1
```

```python
a2 [1 , 0:2] # slice the array to get elements from columns 0 to 1 in row 1
```

```python
a2 [0: , 1:] # slice the array to get all rows and columns from index 1 to the end
```
