```python
import numpy as np
import pandas as pd
```

### Re-Indexing & Altering Labels

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
}

df = pd.DataFrame(data, index=['a', 'b', 'c', 'd'])
df
```

```python
df.reindex(['c', 'd', 'a', 'b'])
```

```python
df.reindex(['a', 'b', 'c', 'd', 'e', 'f'])
```

```python
df
```

### Rename index and columns

```python
df.rename(index = {'a': 'A', 'b': 'B', 'c': 'C', 'd': 'D'}, columns = {'Name': 'Full Name', 'Age': 'Years'})
df
```

### Directly Replace Index or columns Labels

```python
df.index = ['A', 'B', 'C', 'D']
df.columns = ['PersonName', 'PersonAge']
```
