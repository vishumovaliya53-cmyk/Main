### Broadcasting of Array

#### Broadcasting is a Powerful machansim in numpy that allows array operations between arrays of Different Shape without copying data.

#### in simple terms:- Broadcasting allows to you automatically expand smaller arrays to match the shape of larger arrays during operations  like addition , multiplication , etc


### Types of Broadcasting

#### 1.Scaler Broadcasting
#### 2.Vactor Broadcasting
#### 3.Matrix Broadcasting

## # 1.Scaler Broadcasting [e.g. Single Element Array]

```python

 arr
```

```python
# Scaler -> one Element array
s1 = [3]
s2 = 5
s3 = np.array([6])
```

```python
arr + s1  # Broadcasting s1 to match arr shape
```

```python
arr + s2 # Broadcasting s2 to match arr shape
```

```python
arr + s3 # Broadcasting s3 to match arr shape
```

## 2.Vactor Broadcasting

```python
a1 = np.array([[1, 2, 3], 
               [4, 5, 6]])

a2 = np.array([[7, 8, 9]])

print(a1 + a2) 
```

```python
a1 = np.array([[1, 2, 3],
               [4, 5, 6]])

a2 = np.array([10, 20]).reshape(-1, 1)

print(a1 + a2)

a2.shape = (2, 1)
```

## 3.Matrix Broadcasting

```python
a1 = np.array([[1] , [2]]) # 2x1 array
a2 = np.array([[10, 20, 30]]) # 1x3 array

print(a1) 
print()
print(a2)
print()

a3 = a1 + a2 # Broadcasting to 2x3 array
print(a3) 
```
