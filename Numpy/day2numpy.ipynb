## Methemetical Operation

```python
import numpy as np
a1 = np.array([1 , 5 , 7 , 3 , 5 , 2 ]) # create a 1D numpy array
a1
```

```python
a2 = ([2 , 4 , 5 , 9 , 8 , 3]) # create a 2D numpy array
a2
```

```python
a1 + a2 # element-wise addition of two arrays
```

```python
a1 - a2 # element-wise subtraction of two arrays
```

```python
a1 * a2 # element-wise multiplication of two arrays
```

```python
a1 / a2 # element-wise division of two arrays
```

```python
ans = a1 ** 2 # element-wise square of the array
```

```python
np.sqrt(ans)   # both are same sqrt / **2
```

```python
a3 = np.array([1.2 , 2.5 , 3.7 , 4.9]) # create a 1D numpy array with float values
a3
```

```python
np.floor(a3) # floor value
```

```python
np.ceil(a3) # ceiling value
```

```python
np.round(a3) # round off value
```
