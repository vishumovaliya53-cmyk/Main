```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1.Bar Plot - Titanic Dataset 

```python
# load dataset
df = sns.load_dataset("titanic")

# bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='class', y='fare', data=df, estimator=np.sum, errorbar=None, color='lightblue')
plt.xlabel('Passenger Class')
plt.ylabel('Total Fare')
plt.title('Total Fare by Passenger Class')
plt.show()
```

```python
# real word styled dataset
df = pd.DataFrame({
    'Category': ['Laptop', 'Laptop', 'Laptop', 'phone', 'phone', 'Tablet', 'Tablet', 'Tablet'],
    'rating' : [4.5, 4.0, 4.7, 3.8, 4.1, 4.3, 4.6, 3.9]
})

#Bar plot
sns.barplot(x='Category', y='rating', data=df, errorbar=None, palette='pastel', hue='Category')
plt.title('Average Rating by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Rating')
plt.show()
```

```python
df = pd.DataFrame({
    'Category': ['Laptop', 'Laptop', 'phone', 'phone', 'Tablet', 'Tablet'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'rating' : [4.5, 4.0, 3.8, 4.1, 4.3, 4.6]
})

sns.barplot(x='Category', y='rating', data=df, errorbar=None, palette='pastel', hue='Gender')
plt.title('Average Rating by Product Category and Gender')
plt.xlabel('Product Category')
plt.ylabel('Average Rating')
plt.show()
```

## 2.Line Plot - Flight Dataset

```python
flights = sns.load_dataset("flights")
line_data = flights.groupby('year')["passengers"].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='passengers', data=line_data, marker='o')
plt.xlabel('Year')
plt.ylabel('Total Passengers')
plt.title('Total Passengers Over Years')
plt.show()
```

## 3.Scatter plot - Tips Dataset

```python
tips = sns.load_dataset("tips")

plt.figure(figsize=(10, 6))
# sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
sns.regplot(data=tips, x="total_bill", y="tip", scatter =True, color='lightblue')
plt.title('Scatter Plot of Tips vs Total Bill by Day')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

## 4.Histrogram Plot - Titanic 

```python
titanic = sns.load_dataset("titanic")

plt.figure(figsize=(8, 5))
sns.histplot(titanic['age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## 5.Count Plot - Titanic Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
titanic = sns.load_dataset("titanic")

plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x='class', hue='sex', palette='pastel')
plt.title('Count of Reviews by Product Category and Gender')
plt.xlabel('Product Category')
plt.ylabel('Count of Reviews')
plt.show()
```

## 6.Pie Plot - Titanic Dataset

```python
titanic = sns.load_dataset("titanic")

# count survived vs not survived
survival_counts = titanic['survived'].value_counts()

# labels convert: 0 = Not Survived, 1 = Survived
labels = ['Not Survived', 'Survived']

# pie chart
plt.figure(figsize=(8, 8))
plt.pie(survival_counts, labels=labels, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Survival Distribution (Survived vs Not Survived)')
plt.show()
```

## 7.Heatmap Plot - Titanic Dataset --> Show Correlation

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
corr = tips.corr(numeric_only=True)

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, linewidths=0.5, cmap='Pastel2')
plt.title('Correlation Heatmap of Tips Dataset')
plt.show()
```
